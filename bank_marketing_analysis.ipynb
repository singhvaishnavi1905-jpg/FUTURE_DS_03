```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('bank-additional-full.csv', sep = ';')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df['y'].value_counts()
df['y'].value_counts(normalize=True)*100
```

```python
sns.countplot(x='y', data=df, palette='Set2')
plt.title('Target Variable Distribution (Subscribed vs Not)')
plt.xlabel('Subscribed (y)')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('target_distribution.png', dpi=150)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['age'], bins=30, kde=True, color='steelblue')
plt.title('Age Distribution of Clients')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('age_distribution.png', dpi=150)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='job', hue='y', data=df, palette='Set2')
plt.title('Subscription Rate by Job Type')
plt.xlabel('Job')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('job_subscription.png', dpi=150)
plt.show()
```

```python
month_order = ['jan','feb','mar','apr','may','jun',
               'jul','aug','sep','oct','nov','dec']

plt.figure(figsize=(12,5))
sns.countplot(x='month', hue='y', data=df, 
              order=month_order, palette='Set2')
plt.title('Subscription Rate by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('month_subscription.png', dpi=150)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='y', y='duration', data=df, palette='Set2')
plt.title('Call Duration vs Subscription Outcome')
plt.xlabel('Subscribed (y)')
plt.ylabel('Call Duration (seconds)')
plt.tight_layout()
plt.savefig('duration_subscription.png', dpi=150)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), 
            annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=150)
plt.show()
```

```python
df.to_csv('bank_marketing_cleaned.csv', index=False)
print("File saved successfully!")
```
